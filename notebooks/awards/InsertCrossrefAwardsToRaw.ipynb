### Insert Crossref awards (provenance='crossref_work', priority=1) into openalex_awards_raw

Reads from `openalex.awards.crossref_awards` (materialized by the CreateCrossrefAwards DLT pipeline) and refreshes the priority=1 slice of `openalex.awards.openalex_awards_raw`.

Run downstream of the CreateCrossrefAwards DLT pipeline (Crossref_Awards task in CreateWorkAwards job).

```python
%sql
-- Remove previous data for this source before inserting fresh data
DELETE FROM openalex.awards.openalex_awards_raw
WHERE provenance = 'crossref_work' AND priority = 1;

-- Insert into openalex_awards_raw with priority (run after DLT pipeline completes)
INSERT INTO openalex.awards.openalex_awards_raw
SELECT
    id,
    display_name,
    description,
    funder_id,
    funder_award_id,
    amount,
    currency,
    funder,
    funding_type,
    funder_scheme,
    provenance,
    start_date,
    end_date,
    start_year,
    end_year,
    lead_investigator,
    co_lead_investigator,
    investigators,
    landing_page_url,
    doi,
    works_api_url,
    created_date,
    updated_date,
    1 as priority
FROM openalex.awards.crossref_awards
WHERE openalex.common.is_usable_award_id(funder_award_id);
```
