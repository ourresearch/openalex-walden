# Create BBVA Foundation Frontiers of Knowledge Awards (PRIZE PATTERN)

BBVA Foundation Frontiers of Knowledge laureates from the official English laureates page and detail pages.

**Prerequisite:** run `scripts/local/bbva_frontiers_to_s3.py` to fetch the official laureates page, verify the official conditions/amount rule, fetch official detail pages, and upload parquet to S3.

**Source:** `https://www.frontiersofknowledgeawards-fbbva.es/laureates/` plus per-laureate official detail pages; amount-rule source is `https://www.frontiersofknowledgeawards-fbbva.es/conditions/`.

**S3 location:** `s3://openalex-ingest/awards/bbva_frontiers/bbva_frontiers_laureates.parquet`

**OpenAlex funder:** Fundación BBVA, `funder_id = 4320309764`, DOI `10.13039/100007406`; no ROR currently exposed by OpenAlex.

**Schema notes:**
- Prize pattern: one row per official laureate card.
- `lead_investigator` is the laureate. Some official laureates are organizations/groups; these preserve the full official laureate name with `given_name = NULL` and `family_name = laureate_name` because the awards schema is person-shaped.
- `funding_type = 'prize'`.
- Provenance is `bbva_frontiers`.
- The official conditions page states each category carries EUR 400,000, divided when the award is shared. `amount` is therefore `400000 / count(laureates in edition/category)` and `currency = 'EUR'`.
- The source does not publish affiliation-when-awarded as a separate structured field, so `lead_investigator.affiliation` is NULL rather than inferred from biographies.


## Step 1: Create Raw Table

```python
%sql
CREATE OR REPLACE TABLE openalex.awards.bbva_frontiers_raw
USING delta AS
SELECT *, current_timestamp() AS databricks_ingested_at
FROM parquet.`s3a://openalex-ingest/awards/bbva_frontiers/bbva_frontiers_laureates.parquet`;
```

## Step 1.6: Funder Existence Fail-Fast

Must return exactly one row. If zero rows, stop before transforming because the `CROSS JOIN` would emit zero awards.


```python
%sql
SELECT funder_id, display_name, ror_id, doi
FROM openalex.common.funder
WHERE funder_id = 4320309764;
```

## Step 2: Transform to OpenAlex Awards Schema

```python
%sql
CREATE OR REPLACE TABLE openalex.awards.bbva_frontiers_awards
USING delta
AS
WITH funder_resolved AS (
    SELECT funder_id, display_name, ror_id, doi
    FROM openalex.common.funder
    WHERE funder_id = 4320309764
),
normalized AS (
    SELECT
        n.*,
        TRY_CAST(n.award_year AS INT) AS award_year_int,
        TRY_CAST(n.source_award_amount AS DOUBLE) AS amount_double
    FROM openalex.awards.bbva_frontiers_raw n
),
awards AS (
    SELECT
        abs(xxhash64(CONCAT(
            TRY_CAST(f.funder_id AS STRING), ':bbva-frontiers:', LOWER(n.funder_award_id)
        ))) % 9000000000 AS id,
        CONCAT('BBVA Foundation Frontiers of Knowledge Award ', TRY_CAST(n.award_year_int AS STRING), ' - ', n.award_category, ' - ', n.laureate_name) AS display_name,
        COALESCE(NULLIF(n.award_summary, ''), NULLIF(n.contribution, ''), NULLIF(n.bio, '')) AS description,
        f.funder_id,
        n.funder_award_id,
        n.amount_double AS amount,
        NULLIF(n.currency, '') AS currency,
        struct(
            CONCAT('https://openalex.org/F', TRY_CAST(f.funder_id AS STRING)) AS id,
            f.display_name,
            f.ror_id,
            f.doi
        ) AS funder,
        'prize' AS funding_type,
        NULLIF(n.award_category, '') AS funder_scheme,
        'bbva_frontiers' AS provenance,
        TRY_TO_DATE(CONCAT(TRY_CAST(n.award_year_int AS STRING), '-01-01'), 'yyyy-MM-dd') AS start_date,
        TRY_TO_DATE(CONCAT(TRY_CAST(n.award_year_int AS STRING), '-12-31'), 'yyyy-MM-dd') AS end_date,
        n.award_year_int AS start_year,
        n.award_year_int AS end_year,
        struct(
            NULLIF(n.given_name, '') AS given_name,
            NULLIF(n.family_name, '') AS family_name,
            CAST(NULL AS STRING) AS orcid,
            CAST(NULL AS DATE) AS role_start,
            CAST(NULL AS STRUCT<name:STRING, country:STRING, ids:ARRAY<STRUCT<id:STRING, type:STRING, asserted_by:STRING>>>) AS affiliation
        ) AS lead_investigator,
        CAST(NULL AS STRUCT<
            given_name:STRING, family_name:STRING, orcid:STRING, role_start:DATE,
            affiliation:STRUCT<name:STRING, country:STRING, ids:ARRAY<STRUCT<id:STRING, type:STRING, asserted_by:STRING>>>
        >) AS co_lead_investigator,
        CAST(NULL AS ARRAY<STRUCT<
            given_name:STRING, family_name:STRING, orcid:STRING, role_start:DATE,
            affiliation:STRUCT<name:STRING, country:STRING, ids:ARRAY<STRUCT<id:STRING, type:STRING, asserted_by:STRING>>>
        >>) AS investigators,
        NULLIF(n.landing_page_url, '') AS landing_page_url,
        CAST(NULL AS STRING) AS doi,
        current_timestamp() AS created_date,
        current_timestamp() AS updated_date
    FROM normalized n
    CROSS JOIN funder_resolved f
    WHERE n.funder_award_id IS NOT NULL
      AND n.award_year_int IS NOT NULL
      AND n.laureate_name IS NOT NULL
)
SELECT
    id,
    display_name,
    description,
    funder_id,
    funder_award_id,
    amount,
    currency,
    funder,
    funding_type,
    funder_scheme,
    provenance,
    start_date,
    end_date,
    start_year,
    end_year,
    lead_investigator,
    co_lead_investigator,
    investigators,
    landing_page_url,
    doi,
    CONCAT('https://api.openalex.org/works?filter=awards.id:G', TRY_CAST(id AS STRING)) AS works_api_url,
    created_date,
    updated_date
FROM awards;
```

## Step 3: Insert Into `openalex_awards_raw` With Priority 66

```python
%sql
DELETE FROM openalex.awards.openalex_awards_raw
WHERE provenance = 'bbva_frontiers' AND priority = 66;

INSERT INTO openalex.awards.openalex_awards_raw
SELECT
    id,
    display_name,
    description,
    funder_id,
    funder_award_id,
    amount,
    currency,
    funder,
    funding_type,
    funder_scheme,
    provenance,
    start_date,
    end_date,
    start_year,
    end_year,
    lead_investigator,
    co_lead_investigator,
    investigators,
    landing_page_url,
    doi,
    works_api_url,
    created_date,
    updated_date,
    66 AS priority
FROM openalex.awards.bbva_frontiers_awards;
```
