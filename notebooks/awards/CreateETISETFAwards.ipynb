# Create Estonian Science Foundation (ETF) Awards (ETIS)

**Estonian Science Foundation (ETF)** (funder_id `4320321018`, in common.funder, priority `365`). Source: Estonia's
national research-info system ETIS — public REST/JSON API `etis.ee:7443/api/project/getitems`
(see scripts/local/etis_to_s3.py), SPLIT by financier from the 26,265-project national DB
to attribute correctly (FinancingInstitutions = Estonian Science Foundation). **~2078 projects**, native FinancierProjectNr ids,
~100% title/PI/institution/EUR-amount/dates/scheme, bilingual (English fields used).
ETIS aggregates all Estonian funders — ETF (F4320321018) and ETAg (F4320321090) are
SEPARATE OpenAlex entities, attributed here by FinancingInstitutions.

Parquet: `s3://openalex-ingest/awards/etis_etf/etis_etf_grants.parquet`

## Step 1: Create Staging Table from S3

```python
%sql
CREATE OR REPLACE TABLE openalex.awards.etis_etf_raw
USING delta
AS
SELECT *, current_timestamp() as databricks_ingested_at
FROM parquet.`s3a://openalex-ingest/awards/etis_etf/etis_etf_grants.parquet`;
```

```python
%sql
SELECT COUNT(*) as total_projects FROM openalex.awards.etis_etf_raw;
```

## Step 2: Create Awards Table

```python
%sql
CREATE OR REPLACE TABLE openalex.awards.etis_etf_awards
USING delta
AS
WITH
fnd AS (
    SELECT CAST(funder_id AS BIGINT) as funder_id, display_name, ror_id, doi
    FROM openalex.common.funder WHERE funder_id = 4320321018
),
awards_transformed AS (
    SELECT
        abs(xxhash64(CONCAT(f.funder_id, ':', LOWER(g.funder_award_id)))) % 9000000000 as id,
        COALESCE(NULLIF(TRIM(g.title), ''), CONCAT('Estonian Science Foundation (ETF) grant ', g.funder_award_id)) as display_name,
        g.description as description,
        f.funder_id,
        g.funder_award_id as funder_award_id,
        CASE WHEN TRY_CAST(g.amount AS DECIMAL(18,2)) > 0 THEN TRY_CAST(g.amount AS DECIMAL(18,2)) ELSE NULL END as amount,
        CASE WHEN TRY_CAST(g.amount AS DECIMAL(18,2)) > 0 THEN g.currency ELSE NULL END as currency,
        struct(CONCAT('https://openalex.org/F', f.funder_id) as id, f.display_name, f.ror_id, f.doi) as funder,
        'grant' as funding_type,
        g.scheme as funder_scheme,
        'etis_etf' as provenance,
        TRY_TO_DATE(g.start_date_raw, 'yyyy-MM-dd') as start_date,
        TRY_TO_DATE(g.end_date_raw, 'yyyy-MM-dd') as end_date,
        YEAR(TRY_TO_DATE(g.start_date_raw, 'yyyy-MM-dd')) as start_year,
        YEAR(TRY_TO_DATE(g.end_date_raw, 'yyyy-MM-dd')) as end_year,
        CASE WHEN g.pi_family IS NOT NULL THEN
            struct(g.pi_given as given_name, g.pi_family as family_name, CAST(NULL AS STRING) as orcid, CAST(NULL AS DATE) as role_start,
                struct(g.institution as name, 'Estonia' as country, CAST(NULL AS ARRAY<STRUCT<id:STRING, type:STRING, asserted_by:STRING>>) as ids) as affiliation)
        ELSE NULL END as lead_investigator,
        CAST(NULL AS STRUCT<given_name:STRING, family_name:STRING, orcid:STRING, role_start:DATE, affiliation:STRUCT<name:STRING, country:STRING, ids:ARRAY<STRUCT<id:STRING, type:STRING, asserted_by:STRING>>>>) as co_lead_investigator,
        CAST(NULL AS ARRAY<STRUCT<given_name:STRING, family_name:STRING, orcid:STRING, role_start:DATE, affiliation:STRUCT<name:STRING, country:STRING, ids:ARRAY<STRUCT<id:STRING, type:STRING, asserted_by:STRING>>>>>) as investigators,
        g.landing_page_url,
        CAST(NULL AS STRING) as doi,
        CAST(NULL AS STRING) as works_api_url,
        current_timestamp() as created_date,
        current_timestamp() as updated_date
    FROM openalex.awards.etis_etf_raw g
    CROSS JOIN fnd f
)
SELECT * FROM awards_transformed;
```

## Step 3: Insert into openalex_awards_raw

```python
%sql
DELETE FROM openalex.awards.openalex_awards_raw WHERE provenance = 'etis_etf' AND priority = 365;
INSERT INTO openalex.awards.openalex_awards_raw
SELECT id, display_name, description, funder_id, funder_award_id, amount, currency, funder, funding_type, funder_scheme, provenance, start_date, end_date, start_year, end_year, lead_investigator, co_lead_investigator, investigators, landing_page_url, doi, works_api_url, created_date, updated_date, 365 as priority
FROM openalex.awards.etis_etf_awards;
```

## Verification

```python
%sql
SELECT COUNT(*) total, COUNT(DISTINCT funder_award_id) uniq_award, COUNT(DISTINCT id) uniq_id, COUNT(DISTINCT funder_id) funders,
  SUM(CASE WHEN display_name IS NULL OR LENGTH(TRIM(display_name))=0 THEN 1 ELSE 0 END) blank,
  COUNT(amount) has_amount, COUNT(lead_investigator) has_pi, COUNT(start_date) has_start, ROUND(SUM(amount)/1000000,1) total_meur, MIN(start_year) min_yr, MAX(start_year) max_yr
FROM openalex.awards.etis_etf_awards;
```

```python
%sql
SELECT funder_scheme, COUNT(*) cnt, ROUND(SUM(amount)/1000000,1) meur FROM openalex.awards.etis_etf_awards GROUP BY 1 ORDER BY 2 DESC LIMIT 10;
```
