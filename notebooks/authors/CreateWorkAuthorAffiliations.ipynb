### One time job to copy `openalex.authors.author_institutions` to `openalex.works.work_author_affiliations`

```sql
-- copy old table and rename it, adding new columns, remove once system is migrated
CREATE OR REPLACE TABLE openalex.works.work_author_affiliations
USING DELTA
TBLPROPERTIES (
    'delta.deletedFileRetentionDuration' = 'interval 15 days',
    'delta.logRetentionDuration' = 'interval 15 days'
)
CLUSTER BY (work_id) AS
SELECT 
    work_id,
    author_sequence,
    institution_id,
    -- New Column: Author ID (Initialize as NULL for existing records)
    CAST(NULL AS BIGINT) AS author_id,
    raw_author_name,
    raw_affiliation_string,
    raw_countries,
    -- New Columns: Timestamps (Backfill with current time)
    current_timestamp() AS created_at,
    current_timestamp() AS updated_at
FROM openalex.authors.author_institutions;
```
