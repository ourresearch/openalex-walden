# Materialize xpac Work IDs

One-time snapshot of all xpac work IDs as of 2026-03-10.
This freezes the xpac set so no new works can become xpac.

After running this, CreateWorksBase joins against this table instead of using inline logic.

```sql
CREATE OR REPLACE TABLE openalex.works.xpac_work_ids AS
SELECT id AS work_id
FROM openalex.works.openalex_works
WHERE is_xpac = TRUE
```

```sql
SELECT COUNT(*) as xpac_count FROM openalex.works.xpac_work_ids
```
