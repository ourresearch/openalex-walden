# Vacuum and Optimize Tables

Weekly maintenance job to vacuum Delta tables and optimize clustering.

## Tables
- `openalex.works.openalex_works`
- `openalex.works.openalex_works_base`
- `openalex.works.locations_w_sources`

## Vacuum Retention Policy
Tables are vacuumed matching their configured retention: 30 days (720 hours) for openalex_works and locations_w_sources, 60 days (1440 hours) for openalex_works_base.

## Optimize
OPTIMIZE runs on works tables to maintain liquid clustering for efficient queries. Uses incremental optimization targeting recent data.

```python
%sql
VACUUM openalex.works.openalex_works RETAIN 720 HOURS
```

```python
%sql
VACUUM openalex.works.openalex_works_base RETAIN 1440 HOURS
```

```python
%sql
VACUUM openalex.works.locations_w_sources RETAIN 720 HOURS
```

```python
%sql
-- Optimize for liquid clustering (no predicate needed - handles incrementally)
OPTIMIZE openalex.works.openalex_works
```

```python
%sql
-- Optimize for liquid clustering (no predicate needed - handles incrementally)
OPTIMIZE openalex.works.openalex_works_base
```
