# Vacuum Tables

Weekly maintenance job to vacuum Delta tables and reclaim storage space.

- `openalex.works.openalex_works`
- `openalex.works.openalex_works_base`
- `openalex.works.locations_w_sources`

## Retention Policy
All tables are vacuumed with a 30-day retention (720 hours), preserving time travel capability within that window.
