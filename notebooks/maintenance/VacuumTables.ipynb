# Vacuum Tables

Weekly maintenance job to vacuum Delta tables and reclaim storage space.

## Tables
- `openalex.works.openalex_works`
- `openalex.works.openalex_works_base`
- `openalex.works.locations_w_sources`

## Retention Policy
All tables are vacuumed with a 30-day retention (720 hours), preserving time travel capability within that window.

```python
TABLES_TO_VACUUM = [
    "openalex.works.openalex_works",
    "openalex.works.openalex_works_base",
    "openalex.works.locations_w_sources",
]

RETENTION_HOURS = 720  # 30 days
```

```python
# Check current table properties before vacuum
for table in TABLES_TO_VACUUM:
    print(f"\n{'='*60}")
    print(f"Table: {table}")
    print(f"{'='*60}")
    display(spark.sql(f"DESCRIBE DETAIL {table}"))
```

```python
# Vacuum each table with 30-day retention
for table in TABLES_TO_VACUUM:
    print(f"\nVacuuming {table} with {RETENTION_HOURS} hours retention...")
    spark.sql(f"VACUUM {table} RETAIN {RETENTION_HOURS} HOURS")
    print(f"Completed vacuum for {table}")
```

```python
# Verify table sizes after vacuum
for table in TABLES_TO_VACUUM:
    print(f"\n{'='*60}")
    print(f"Table: {table} (after vacuum)")
    print(f"{'='*60}")
    display(spark.sql(f"DESCRIBE DETAIL {table}"))
```
