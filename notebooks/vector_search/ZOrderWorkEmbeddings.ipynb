# Z-ORDER tables for fast lookups
One-time optimization for work_embeddings_v2 (by work_id) and authorship_embedding_similarity (by author_id).

```python
import time

tables = [
    ("openalex.vector_search.work_embeddings_v2", "work_id"),
    ("openalex.aer.authorship_embedding_similarity", "author_id"),
]

results = []
for table, col in tables:
    t0 = time.time()
    print(f'OPTIMIZE {table} ZORDER BY ({col})...')
    spark.sql(f'OPTIMIZE {table} ZORDER BY ({col})')
    elapsed = (time.time() - t0) / 60
    print(f'  Done in {elapsed:.1f} min')
    results.append(f'{table}: {elapsed:.1f} min')

output = "ZORDER complete: " + ", ".join(results)
print(output)
dbutils.notebook.exit(output)
```
