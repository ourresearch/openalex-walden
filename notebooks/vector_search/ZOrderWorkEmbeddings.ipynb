# Z-ORDER work_embeddings_v2 by work_id
One-time optimization to enable fast lookups by work_id.

```python
import time
t0 = time.time()
print('Starting OPTIMIZE ZORDER...')
spark.sql('OPTIMIZE openalex.vector_search.work_embeddings_v2 ZORDER BY (work_id)')
elapsed = (time.time() - t0) / 60
print(f'Done in {elapsed:.1f} min')
dbutils.notebook.exit(f'ZORDER complete in {elapsed:.1f} min')
```
