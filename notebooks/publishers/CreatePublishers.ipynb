```sql
CREATE OR REPLACE TABLE openalex.publishers.publishers
AS SELECT 
    p.publisher_id as id,
    p.display_name,
    p.alternate_titles,
    p.country_code,
    CASE 
      WHEN p.parent_publisher IS NOT NULL THEN
        STRUCT(
          CONCAT('https://openalex.org/P', p.parent_publisher) as id,
          parent.display_name as display_name
        )
      ELSE NULL
    END as parent_publisher,
    p.ror_id,
    p.wikidata_id,
    p.country_name,
    p.parent_publisher_wikidata,
    p.hierarchy_level,
    p.country_codes,
    p.image_url,
    p.image_thumbnail_url,
    p.homepage_url,
    p.merge_into_id,
    p.merge_into_date,
    p.created_date,
    p.created_date as updated_date
FROM openalex_postgres.mid.publisher p
LEFT JOIN openalex.publishers.publishers parent 
  ON p.parent_publisher = parent.id AND parent.merge_into_id IS NULL
```
