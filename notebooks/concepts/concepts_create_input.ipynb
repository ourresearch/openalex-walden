```python
%sql
-- Populate concepts input table with works that need concept inference
-- Excludes works already in work_concepts (via LEFT ANTI JOIN by work_id)
CREATE OR REPLACE TABLE openalex.works.work_concepts_input
SELECT
  id AS work_id,
  title,
  abstract,
  primary_location.source.display_name AS journal,
  primary_location.source.type AS doc_type
FROM openalex.works.openalex_works_base w
LEFT ANTI JOIN openalex.works.work_concepts wc
  ON w.id = wc.work_id
WHERE title IS NOT NULL
  AND ((length(title) > 20 AND length(abstract) > 50)
       OR length(title) > 50
       OR length(abstract) > 150);
```

```python
%sql
OPTIMIZE openalex.works.work_concepts_input ZORDER BY (work_id);
```

```python
%sql
SELECT FORMAT_NUMBER(COUNT(*),0) as num_works_to_process
FROM openalex.works.work_concepts_input
```
